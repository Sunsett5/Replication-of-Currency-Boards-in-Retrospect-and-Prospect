```python
import pandas as pd
import datetime as dt
import numpy as np
import statsmodels.tsa.filters.hp_filter as hpfilter
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
start_year = 1970
end_year = 2023
year_obs = end_year - start_year +1
```

```python
dfWEObase = pd.read_excel('WEO-base.xlsx',index_col = 0)
WEO_list = sorted(list(set(dfWEObase['NAM'])))
dfWEObase
```

```python
dfIFSbase = pd.read_excel('IFS-base.xlsx',index_col = 0)
dfIFSbase
```

```python
dfWolfbase  = pd.read_excel('Wolf-base.xlsx',index_col = 0)
Wolf_list = sorted(list(set(dfWolfbase['NAM'])))
dfWolfbase
```

```python
dictWolf = {}
dfWolfextended = None
year_list = list(range(start_year, end_year+1))*len(WEO_list) 
country_list = []
for country in WEO_list:
    country_list += [country]*year_obs
    if country in Wolf_list:
        dictWolf[country] = pd.concat([dfWolfbase.loc[dfWolfbase['NAM']==country,:],
                                       pd.DataFrame(np.nan, index=list(range(2003,end_year+1)),columns=dfWolfbase.columns)],axis=0)
    else:
        dictWolf[country] = pd.DataFrame(np.nan, index=list(range(start_year,end_year+1)),columns=dfWolfbase.columns)
    if country == 111:
        dfWolfextended = dictWolf[country]
    else:
        dfWolfextended = pd.concat([dfWolfextended,dictWolf[country]],axis=0)

dfWolfextended['YRS'] = year_list
dfWolfextended['NAM'] = country_list
dfWolfextended.reset_index(drop = True, inplace = True)
        
dfWolfextended
```

```python
dfWEObase.update(dfIFSbase,overwrite = False)
dfWEObase.update(dfWolfextended,overwrite = False)
dfWEObase
```

```python
dfWEObase.to_excel('CombinedData.xlsx')
```
